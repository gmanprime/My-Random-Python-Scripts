```python
from pprint import pprint as pp
import fiona  # ? not sure what this does ?
import geopandas as gpd  # geo analysis toolbox based on pandas data analysis toolbox
import os  # used for accessing Operating system functionality, in this case, the path computation tools
import argparse  # can be used to get arguments trailing the python call when running code
```

Next we define the path where the geopackage is located

```python
# test file path to GeoPackage
geoPath = os.path.join(os.path.expanduser("~"), "Documents", "AnalysisPackages", "customerCompare.gpkg")
```

we then define variables for Manipulation

```python
layers = fiona.listlayers(geoPath)
geoData = {}
customerKeys = {}
```

we then define a function that handles the compare operation

```python
def idMatch():
        # This function compares the specified id value from an attribute
        for layer in layers:
            # read the data from the geopackage file, and specifically a layer from the fiona module
            geoData[layer] = gpd.read_file(geoPath, layer=layer).to_dict(orient='records')

            # set the layer array list of customers to empty array
            customerKeys[layer] = {}

            # add the customers to the layers array list
            for recordDict in geoData[layer]:
                if ("CUSTOMERKE" not in recordDict):
                    # add the customer key to the dict as a key with value of True
                    customerKeys[layer][recordDict['CUST_KEY']] = True
                elif ("CUSTOMERKE" in recordDict):
                    # add the customer key to the dict as a key with value of True
                    customerKeys[layer][recordDict['CUSTOMERKE']] = True
                    

        
```

once we have all the customer keys, we now filter out the bool values in order to get compute the number of matching values

```python
def filter():
   # get the list of customer keys from first dictionary
   for key, value in customerKeys[layers[0]].items():
      if key in customerKeys[layers[1]].keys(): # if the key value exists
         Match = value == customerKeys[layers[1]][key] # compare the values and generate a match bool
         
         if Match: # if the match bool is true
            customerKeys[layers[0]][key] = False #set the customer value to false that didnt have a match
            customerKeys[layers[1]]
```

we now do a compare function to calculate the efficiency

```python
def compare():
   countMatching = {} #............................................# cKey count values for both layers
   for layer in layers: #..........................................# do below actions for both layers
      countMatching[layer] = 0 #...................................# reset the final count value to 0
      for TBool in customerKeys[layer].values(): #.................# for Truth value in customerKeys
         if TBool == True: #.......................................# if the value is True
            countMatching[layer]+=1 #..............................# add to the count matching var for that layer
   return countMatching

print(compare()[layers[0]], ' ', len(customerKeys[layers[0]]))#....# take the count of True values and substract it from the total amount of values available
print(len(customerKeys[layers[0]]) - compare()[layers[0]])
print(compare()[layers[1]], ' ', len(customerKeys[layers[1]]))#....# take the count of True values and substract it from the total amount of values available
print(len(customerKeys[layers[1]]) - compare()[layers[1]])
```
